```python
import requests
import pandas as pd
```

```python
def fetch_current_weather(city, api_key):
    url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    data = requests.get(url)
    
    if data.status_code==200:
        print("page fetched successfully done")
    elif data.status_code==404:
        print("city is not Found")
    elif data.status_code==429:
        print("API rate limit exceeded - too many requests")
    else:
        print("failed to fetch page")
    
    return data.json()

def fetch_5day_forecast(city_name, api_key):
    url = f"https://api.openweathermap.org/data/2.5/forecast?q={city_name}&appid={api_key}&units=metric"
    data = requests.get(url)
    
    if data.status_code==200:
        print("page fetched successfully done")
    elif data.status_code==404:
        print("city is not Found")
    elif data.status_code==429:
        print("API rate limit exceeded - too many requests")
    else:
        print("failed to fetch page")
    
    return data.json()

api_key = "080ca7d6b2cd7b267cda1d2b1467c121"
cities = ["London", "Paris", "Tokyo", "New York","Mumbai","Pune","Delhi","Banglore"]

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

def input_city():
    
    while True:
        A=input("Enter the city that you want to know weather about it [London,Paris,Tokyo,New York,Mumbai,Pune,Delhi,Banglore] choose from the bracket list and write [exit] to exit:")
        city_found = False
        if A.lower() =="exit" :
             print("--Exiting--") 
             break
        while True:
            param=["C","F"]
            B=input("What type of parameters you want in your city temp[C or F] choose from bracket onlyand write [exit] for exit").upper()
            
            if B in param:
                break
            else:
                print("Invalid parameter.Please provide from given Bracket")
                
        for city in cities:
            if(A==city.lower()):
                print(f"Fetching weather for: {city}")
                
                current = fetch_current_weather(city, api_key)
                forecast = fetch_5day_forecast(city, api_key)
                print("")
                print("------------------------------------")
                print("Current weather  :")
                if 'main' in current:
                    temp_c = current['main']['temp']
                    feels_like=current["main"]["feels_like"]
                    if B == 'C':
                        print(f"Temperature: {temp_c}°C")
                        print(f"Temperature feels like in {city} is :{feels_like}°C")
                    else:
                        temp_f = celsius_to_fahrenheit(temp_c)
                        feels_f=celsius_to_fahrenheit(feels_like)
                        print(f"Temperature: {temp_f}°F")
                        print(f"Temperature feels like in {city} is :{feels_f}°F")
                print(f"The humidity in {city} is {current["main"]["humidity"]}%")
                print(f"The condition of weather city {city} is {current['weather'][0]['description']} with icon is {current['weather'][0]['icon']}") 
                print("")
                print("----5 day forecast----")
                
                print("")


                if 'list' in forecast:
                     unique_dates = set()
    
                     for forecast_item in forecast['list']:
                        date_only = forecast_item['dt_txt'].split()[0]  
        
                        if date_only not in unique_dates:
                             unique_dates.add(date_only)
                             print(f"Day {len(unique_dates)}: {date_only}")
            
                             if len(unique_dates) >= 5:
                               break
                             forecast_main = forecast["list"][0]["main"]
                             print(f"Minimum temperature - {forecast_main['temp_min']}°C")
                             print(f"Maximum temperature - {forecast_main['temp_max']}°C")
                             print(f"Pressure - {forecast_main['pressure']}")
                             print(f"sea level- {forecast_main['sea_level']}")
                             print(f"Ground level - {forecast_main['grnd_level']}")
                             print(f"Humidity - {forecast_main['humidity']}")
                             for forecast_weather in forecast["list"][0]["weather"]:
                                 print("----weather----")
                                 print(f"main - {forecast_weather["main"]}")
                                 print(f"Description - {forecast_weather["description"]}")
                                 print(f"icon - {forecast_weather["icon"]}")
                                 print("")
                                 print("-----------------------------------")
                print("Current Weather:", current)
                print("5-Day Forecast:", forecast)
                city_found=True
                break
        if city_found:
            break
        else:
            print("The city is not included.please give the valid city name")
 

   
input_city()
```
